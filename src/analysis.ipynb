```python
import pandas as pd
import os
```

```python
transaction_set_filepath = os.path.join("..", "data", "credit_transactions-ibm_vs.csv")

df = pd.read_csv(transaction_set_filepath)
df.head
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

value_counts = df['Merchant Name'].value_counts()

top_values = value_counts[:20]
other_sum = value_counts[20:].sum()

top_values['Other'] = other_sum

plt.figure(figsize=(6, 6))
plt.pie(top_values, autopct='%1.1f%%', startangle=140)
plt.title('Merchant Distribution')
plt.show()
```
